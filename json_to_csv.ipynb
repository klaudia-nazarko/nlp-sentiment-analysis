# CSV Preparation

Dataset containing Amazon reviews from Toys and Games category is provided in json format. The simple code below will load data, process it and save review and sentiment column in csv format.

Sentiment was assigned based on rating - only highly polar ratings were used. Reviews with rating 5 were labaled as positive, while reviews with 1 and 2 score were categorized as negative. Group of negative reviews contains both 1 and 2 in order to decrease class imbalance.

```python
import pandas as pd
```

```python
file_reviews = 'data/Toys_and_Games_5.json'
df = pd.read_json(file_reviews, orient='records', lines=True)
df.sample(5)
```

```python
df_review = df[['reviewText', 'overall']][df['overall'].isin([1,2,5])]
df_review.reset_index(inplace=True, drop=True)
df_review.columns = ['review', 'rating']
df_review['sentiment'] = df_review['rating'].map({5:1,2:0,1:0})
```

```python
df_review[['review', 'sentiment']].to_csv('data/reviews_toys_games.csv', index=False)
```
